```python
#uppgift E

import random 
import math
import numpy as np

p = 0  #i denna uppgift sak vi räkna punkterna så p=punkt och vi vill att punkterna ska starta på 0 

for i in range(100):
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)
    längd = np.sqrt(x**2+y**2)  
    '''
    denna for sats slumpar en x och en y kordinat mellan -1 och 1 och sedan använder pytagorats sats för att räkna ut avståndet           och repeterar detta 100 gånger
    '''
    if längd < 1: 
        p += 1 #denna if sats räknar punkterna innanför cirkeln AKA om längden är mindre än 1 eftersom radien på cirkeln är 1
    s = p/100 #här delar vi det på totala numret av punkter 
 
print(s)
```

```python


for i in range(1,101):
    if i%5 == 0:# här använder vi en modolus som kollar om i är delbart med 5 och om det är det är det = 0
        print("burr", end = " ")
    else: #om talet inte är delbart med 5 så räknar den som förut igen
        print(i, end = " ") 
        
#end saken är hur man ska skriva efter varje print för vanligtvis är det print() enter och så fortsätter det men nu är det print() space och så fortsätter det så så det sitter ihop mer och man inte behöver skrolla ner 
    
```

```python
x = float(input("Ange ett multiplar som ska bli burr"))# detta är ett input från användaren som vi har bundit till variabeln x

for i in range(1,101):
    if i%x == 0:# här skriver vi så att vilket tal användaren skriver in t,ex 3 så printar den burr på allting som är delbart med 3 
        print("burr", end = " ")
    else: 
        print(i, end = " ")
```

```python
burr = float(input("Ange ett multiplar som ska bli burr"))
birr = float(input("Ange ett multiplar som ska bli birr"))


for i in range(1,101):
    if i%burr == 0 and i%birr == 0:  # denna är här för att om birr och burr blir på samma tal så vill vi att den skriver båda 
        print("birr burr")
    elif i%birr == 0:
        print("birr" )
    elif i%burr == 0:
        print("burr")
    else: 
        print(i, end = " ")

 #denna uppgift är samma princip som förra men det är två variabler som användaren får välja med andra ord birr och burr
```

```python
burr = float(input("Ange ett multiplar som ska bli burr"))
birr = float(input("Ange ett multiplar som ska bli birr"))
# vi gör dessa till int istället för float eftersom for satsen inte kan räkna med en string men kan det med en int vilket gör så att vi kan använda variabler för användarens tal och därför bestämma vart programmet börjar och slutar 
start = int(input("Ange vart programet ska börja"))
end = int(input("Ange vart programmet ska sluta"))

bu = 0  #dessa är variabler för räkning där bu = burr , bi=birr och bibu= birr burr
bi = 0
bibu = 0

for i in range(start,end):
    if i%birr == 0 and i%burr == 0: 
        print("birr burr")
        bibu += 1 # detta använder vi om det blir en birr burr så kan vi räkna det 
    elif i%birr == 0:
        print("birr")
        bi += 1
    elif i%burr == 0:
        print("burr")
        bu += 1
    else: 
        print(i, end = " ")

print(f"Du fick {bi}birr, {bu}burr och {bibu}birr burr")
```
